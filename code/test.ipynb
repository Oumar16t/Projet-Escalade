```python
import random
from datetime import datetime, timedelta
```

```python
# Charger les données météo à partir d'un fichier Excel
weather_data = pd.read_excel("C:\\Users\\oumar\\OneDrive\\Bureau\\Projet_Escalade\\df.xlsx")
weather_data['date'] = pd.to_datetime(weather_data['date'], format='%d/%m/%Y')
weather_data
```

```python
# Fonction pour générer une date aléatoire entre deux dates données
def date_aleatoire(debut, fin):
    # Convertir les dates en objets datetime
    debut_date = datetime.strptime(debut, '%Y-%m-%d')
    fin_date = datetime.strptime(fin, '%Y-%m-%d')
    
    # Calculer la plage de jours entre les dates
    plage_jours = (fin_date - debut_date).days
    
    # Générer un décalage aléatoire dans la plage de jours
    offset_jours = random.randint(0, plage_jours)
    
    # Ajouter le décalage aléatoire à la date de début
    random_date = debut_date + timedelta(days=offset_jours)
    
    return random_date
```

```python
# Fonction pour récupérer les données météo pour une date donnée
def get_weather_by_date(df, date_str):
    date = pd.to_datetime(date_str, format='%d/%m/%Y')
    weather_on_date = df[df['date'] == date]
    if not weather_on_date.empty:
        return weather_on_date
    else:
        print(f"La date {date_str} n'existe pas dans les données.")
        return None


x = get_weather_by_date(weather_data, '14/05/2024')
x
```

```python
# Fonction pour calculer un score météo basé sur différents paramètres météorologiques
def calculate_weather_score(df):
    # Normaliser les paramètres (ajustez les seuils selon vos besoins)
    humidity_score = max(0, min(5, 5 - (float(df['humidity (%)'].str.replace(',', '.').iloc[0]) / 20)))  # Humidité de 0 à 100%
    precipitation_score = max(0, min(5, 5 - (float(df['precipitation (%)'].str.replace(',', '.').iloc[0]) / 20)))  # Précipitations de 0 à 100%
    temp_max_score = max(0, min(5, (float(df['tmp_max (°C)'].str.replace(',', '.').iloc[0]) + 10) / 10))  # Température max de -10°C à 40°C
    temp_min_score = max(0, min(5, (float(df['tmp_min (°C)'].str.replace(',', '.').iloc[0]) + 10) / 10))  # Température min de -10°C à 40°C
    wind_speed_score = max(0, min(5, 5 - (float(df['wind (Km/h)'].str.replace(',', '.').iloc[0]) / 10)))  # Vitesse du vent de 0 à 50 km/h

    # Moyenne pondérée ou simple des notes
    total_score = (humidity_score + precipitation_score + temp_max_score + temp_min_score + wind_speed_score) / 5

    # Arrondir le score final à la note entière la plus proche
    final_score = round(total_score)

    return final_score
```

```python
# Fonction pour créer des blocs de données avec des scores météo
def create_blocks(weather_data, boolder):
    debut = '2024-04-01'
    fin = '2024-05-15'
    blocks = {}
    
    # Boucler sur chaque ligne de la bdd
    for idx, row in boolder.iterrows():
        dates = {}
        
        # Générer 40 entrées de dates et notes pour chaque blocks
        for _ in range(40):
            coin = random.randint(0, 100)
            random_date = date_aleatoire(debut, fin).strftime("%d/%m/%Y")
            
            # Une chance sur trois de générer une note aléatoire
            if (coin % 3 == 0):
                note = random.randint(0, 5)
                dates[random_date] = note
            else:
                note = calculate_weather_score(get_weather_by_date(weather_data, random_date))
                dates[random_date] = note

        name = row['name']
        blocks[row['id']] = [name, dates]
   
    return blocks

blocks = create_blocks(weather_data, boolder)
```

```python
# Fonction pour filtrer les données météo pour un bloc donné et ajouter les scores correspondants
def filtered_weather(id, dico):
    weather_data = pd.read_excel("C:\\Users\\oumar\\OneDrive\\Bureau\\Projet_Escalade\\df.xlsx")
    weather_data['date'] = weather_data['date'].astype(str)
    
    # Obtenir les dates du dictionnaire pour le bloc donné
    dico_dates = list(dico[id][1].keys())
    
    # Filtrer les données météorologiques pour ne conserver que les dates présentes dans le dictionnaire
    filtered_weather_data = weather_data[weather_data['date'].isin(dico_dates)]

    # Extraire les notes du dictionnaire et les convertir en DataFrame
    notes = pd.DataFrame(list(dico[id][1].items()), columns=['date', 'note'])

    # Joindre les notes aux données météo filtrées
    data = pd.merge(filtered_weather_data, notes, on='date')
    
    data.dropna(subset=['note'], inplace=True)
    
    return data
    
data_set = filtered_weather(45, blocks)
```

```python
# Créer un DataFrame pour les nouvelles données
labels = ['date', 'tmp_max (°C)', 'tmp_min (°C)', 'precipitation (%)', 'wind (Km/h)', 'humidity (%)']
df = pd.DataFrame(data,columns=labels)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
```

```python
# Convertir les donné en float 
for column in data_set.columns:
    if data_set[column].dtype == 'object':
        data_set[column] = data_set[column].str.replace(',', '.')
        try:
            data_set[column] = data_set[column].astype(float)
        except ValueError:
            pass  # Ignorer les colonnes qui ne peuvent pas être converties en float

# Séparation des données en entrées (X) et cibles (y)
X = data_set.drop(columns=['note','date','Unnamed: 0'])
y = data_set['note']


# Normalisation des caractéristiques
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Séparation des données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Création d'un modèle de régression k-plus proches voisins (KNN)
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train, y_train) # Entraîne le modèle sur les données d'entraînement

# Prédiction des valeurs cibles pour les données de test
y_pred = knn.predict(X_test)
y_pred_rounded = [round(pred) for pred in y_pred]

# Calcul de l'erreur quadratique moyenne (MSE) pour évaluer les performances du modèle
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred_rounded)
print(f'Mean Squared Error: {mse}')

'''# Calcul de la precision
from sklearn.metrics import precision_score
prs = precision_score(y_test, y_pred, average = None)
print(f'Precision score: {prs}')'''
```
